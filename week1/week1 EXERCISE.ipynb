To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display,Markdown,update_display
```

```python
# constants

MODEL_GEMINI = 'gemini-3-flash-preview'
MODEL_LLAMA = 'gemma3:270m'
```

```python
# set up environment
load_dotenv()

GEMINI_API_KEY=os.getenv("GEMINI_API_KEY")
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
def invoke_llm_gemini(system_prompt,user_prompt):
    gemini = OpenAI(base_url=GEMINI_BASE_URL,api_key=GEMINI_API_KEY)
    stream = gemini.chat.completions.create(model=MODEL_GEMINI,messages=[
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt}
    ],stream=True)
    result=""
    display_handle = display(Markdown(""),display_id=True)
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        update_display(Markdown(result),display_id=display_handle.display_id)

def invoke_llm_ollama(system_prompt,user_prompt):
    ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key="OLLAMA_API")
    response = ollama.chat.completions.create(model=MODEL_LLAMA,messages=[
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt}
    ])
    result= response.choices[0].message.content
    display(Markdown(result))
```

```python
# here is the question; type over this to ask something new

system_prompt = """
you are the Software engineering tutor. who has knowledge about all fields lie around software engineering.
you have to answer user question in funny way, but make sure answer is correct.
if someone ask you question out of your domain than answer ask them to ask question about software engineering only in humourous way.
"""

user_prompt = """
What is RAG and vector Database?
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
invoke_llm_gemini(system_prompt=system_prompt,user_prompt=user_prompt)
```

```python
user_prompt2="who is the president of India?"
invoke_llm_gemini(system_prompt=system_prompt,user_prompt=user_prompt2)
```

```python
# Get Llama 3.2 to answer
user_prompt2="who is the president of america?"
invoke_llm_ollama(system_prompt=system_prompt,user_prompt=user_prompt2)
```
